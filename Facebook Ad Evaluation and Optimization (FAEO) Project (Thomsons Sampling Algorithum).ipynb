**IMPLEMENTATION OF THOMSONS SAMPLING ALGORITHUM**

```python
# WE WILL IMPORT THE LIBRARI HERE
import math
import random

# we will define the total rounds and total adds
T = 15000
num_ads = 10

# we will define the adds selected
ads_selected = []

# for particular add we will define the reward for it
# 0 no of time at each time
num_of_rewards_1 = [0] * num_ads

# same we will do for no of rewards = 0
num_of_rewards_0 = [0] * num_ads

# for knowing the total no of rewards
total_rewards = 0

# we will need 2 for loops for itrating throw all 15000 rounds
# and onther one for each of the advertisement

for n in range(0,T):
  # we will define current add is o
  ad = 0

  # we will create the max_random = we need to select the highest value of this
  max_random = 0

  # now we will create the 2nd for loop
  for i in range(0, num_ads):

    # now we will code for the particular formula that we have
    # we will create the variable for that
    random_beta =  random.betavariate(num_of_rewards_1[i] + 1 , num_of_rewards_0[i] + 1 )

    # we will chek is our random_beta value greater than the max_random value
    # it will obiously greater for the 1 st time
    if (random_beta > max_random):

      # now we will set max_random value as currnet_beta value
      max_random = random_beta

      # we will tell that the ith add selected is
      ad = i
    # we will apppend the add into the ads_selected now
    ads_selected.append(ad)

    # we will call all the dataset values here
    # we want all the values for the n outside for the loop
    # we want this for the particular add
    reward = dataset.values[ n , ad ]

    # now we will say fo the particular add
    if reward == 1 :
      num_of_rewards_1[ad] = num_of_rewards_1[ad] + 1
    else:
      num_of_rewards_0[ad] = num_of_rewards_0[ad] + 1

    # in this for loop we we say the total rewards equal to total_reward + current _reward
    total_rewards = total_rewards + reward
```

**HISTOGRAM FOR FACEBOOK ADS OPTIMIZATION**

```python
plt.hist(ads_selected)
plt.title("HISTOGRAM FOR FACEBOOK ADS OPTIMIZATION AND ANALYSIS")
plt.xlabel("ADS")
plt.ylabel("NO. TIMES EACH ADD WAS SELECTED")
plt.show()
```
